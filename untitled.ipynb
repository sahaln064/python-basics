```python
import pandas as pd
df = pd.read_csv("/content/Student_performance_data _.csv")
print(df.head())
```

```python
print(df.info())
print(df.describe())
```

```python
avg_gpa = df["GPA"].mean()
print("average GPA:",avg_gpa)
```

```python
top_student = df.loc[df["GPA"].idxmax()]
print("top performer:")
print(top_student)
```

```python
df ["Result"]=df["GPA"].apply(lambda x: "pass"if x>=2.5 else "fail")
pass_percentage = (df["Result"] == "pass").mean() * 100
print("pass percentage:",pass_percentage)
```

```python
import pandas as pd
import numpy as np
data = pd.read_csv("/content/train (1).csv")
print(data.head())
```

```python
survived = data['Survived'].to_numpy()
```

```python
print("Survived:",np.sum(survived == 1))
print("not survived:",np.sum(survived == 0))
```
