```python
import pandas as pd
df = pd.read_csv("/content/Student_performance_data _.csv")
print(df.head())
```

```python
print(df.info())
print(df.describe())
```

```python
avg_gpa = df["GPA"].mean()
print("average GPA:",avg_gpa)
```

```python
top_student = df.loc[df["GPA"].idxmax()]
print("top performer:")
print(top_student)
```

```python
df ["Result"]=df["GPA"].apply(lambda x: "pass"if x>=2.5 else "fail")
pass_percentage = (df["Result"] == "pass").mean() * 100
print("pass percentage:",pass_percentage)
```
